```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    {
        'A':[1, np.nan, 3, 4, 5],
        'B':[6, 7, 8, np.nan, 10],
        'C':[11, 12, 13, 14, 15],
    }
)

df
```

```python
df.isnull()
```

```python
from sklearn.impute import SimpleImputer

# 欠損値を埋める種類を設定→１
imp = SimpleImputer(strategy='mean')
# dfの統計値を取得
imp.fit(df)
# dfの欠損値を1で指定したもので埋める
imp.transform(df)
```
