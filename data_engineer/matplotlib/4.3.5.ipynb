```python
import pandas as pd
import matplotlib.style
import matplotlib.pyplot as plt


matplotlib.style.use('ggplot')
```

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B':[3, 1, 2]})
df.plot()

plt.show()
```

```python
import numpy as np

rng = np.random.default_rng(123)
df2 = pd.DataFrame(rng.random((3, 2)),
                   columns=['y1', 'y2'])

df2.plot.bar()
plt.show()
```

```python
df2.plot.bar(stacked=True)
plt.show()
```
