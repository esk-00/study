```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style

matplotlib.style.use('ggplot')
```

```python
rng = np.random.default_rng(123)
mu = 100
sigma = 15
x = rng.normal(mu, sigma, 1000)
fig, ax = plt.subplots()

n, bins, patches = ax.hist(x)

plt.show()
```

```python
for i, num in enumerate(n):
    print(f"{bins[i]:.2f} - {bins[i + 1]:.2f}: {num}")
```

```python
rng = np.random.default_rng(123)
mu = 100
x0 = rng.normal(mu, 20, 1000)

x1 = rng.normal(mu, 15, 1000)
x2 = rng.normal(mu, 10, 1000)

fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']
ax.hist((x0, x1, x2), label=labels)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']
ax.hist((x0, x1, x2), label=labels, stacked=True)
ax.legend()

plt.show()
```

```python
import numpy as np


rng = np.random.default_rng(123)
x = rng.normal(size=1000)

fig, ax = plt.subplots()

counts, edges, patches = ax.hist(x, bins=25)

x_fit = (edges[:-1] + edges[1:]) / 2

y = 1000 * np.diff(edges) * np.exp(-x_fit**2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit, y)

plt.show()
```
