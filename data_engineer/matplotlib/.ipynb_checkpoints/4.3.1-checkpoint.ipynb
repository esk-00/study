```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.style

matplotlib.style.use('ggplot')
```

```python
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y)
plt.title('pyplot interface')

plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 9]

fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title('object-oriented interface')
plt.show()
```
