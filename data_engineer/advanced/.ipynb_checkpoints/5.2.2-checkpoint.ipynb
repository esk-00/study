```python
import MeCab
import ipadic

text = "吾輩は猫である"

CHASEN_ARGS = r'''
 -F "%m\t%f[7]\t%f[6]\t%F-[0,1,2,3]\t%f[4]\t%f[5]\n"
 -U "%m\t%m\t%m\t%F-[0,1,2,3]\t\t\n"
'''

t = MeCab.Tagger(ipadic.MECAB_ARGS + CHASEN_ARGS)
result = t.parse(text)
print(result)
```

```python
import MeCab
import ipadic

documents = ['子供が走る', '車が走る', '子供の脇を車が走る']

words_list = []

CHASEN_ARGS = r'''
 -F "%m\t%f[7]\t%f[6]\t%F-[0,1,2,3]\t%f[4]\t%f[5]\n"
 -U "%m\t%m\t%m\t%F-[0,1,2,3]\t\t\n"
'''

t = MeCab.Tagger(ipadic.MECAB_ARGS + CHASEN_ARGS)

for s in documents:
    s_parsed = t.parse(s)
    words_s = []
    for line in s_parsed.splitlines()[:-1]:
        words_s.append(line.split('\t')[0])
    words_list.append(words_s)

print(words_list)
```

```python
word2int = {}

i = 0

for words in words_list:
    for word in words:
        if word not in word2int:
            word2int[word] = i
            i += 1
        
print(word2int)
```

```python
import numpy as np

bow  = np.zeros((len(words_list), len(word2int)),
               dtype=int)

for i, words in enumerate(words_list):
    for word in words:
        bow[i, word2int[word]] += 1
print(bow)
    
```

```python
import pandas as pd
pd.DataFrame(bow, columns=list(word2int))
```
