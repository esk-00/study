```python
import requests

r = requests.get('https://www.seshop.com/')
print(type(r))
print(r.status_code)
```

```python
text = r.text
for line in text.split('\n'):
    if '<title>' in line or '<h2>' in line:
        print(line.strip())
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(text, 'html.parser')
print(soup.title)
print(soup.h2)
print(soup.h2.a["href"])
```

```python
atags = soup.find_all('a')
print('aタグの数：', len(atags))
for atag in atags[:5]:
    print('タイトル：', atag.text)
    print('リンク：', atag['href'])
```

```python
from datetime import datetime

import requests
from bs4 import BeautifulSoup

r = requests.get('https://www.seshop.com/prodcut/1')
soup = BeautifulSoup(r.text, "html.parser")

books = []

divs = soup.select("div.list div.inner")
for div in divs:
    img_url = div.find("img")["src"]
    day = div.find("span", class_="date").text.strip()
    day = day.replace("発売", "")
    published = datetime.strptime(day, "%Y.%m.%d")
    
    div_txt = div.find("div", class_="txt")
    a_tag = div_txt.find("a")
    title = a_tag.text.strip()
    url = a_tag["href"]
    
    price_s = div_txt.find("p", class_="price").text
    price_s = price_s.strip()
    price_s = price_s.replace("販売価格：", "")
    price_s = price_s.replace("円（税込み）", "")
    price_s = price_s.replace(",", "")
    price = int(price_s)
    
    book = {
        "title": title,
        "img_url": img_url,
        "url": url,
        "price": price,
        "published": published,
    }
    books.append(book)
```
